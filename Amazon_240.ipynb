```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
HEADERS = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/143.0.0.0 Safari/537.36 Edg/143.0.0.0"
    ),
    "Accept-Language": "en-IN,en;q=0.9"
}
```

```python
urls = [
    f"https://www.amazon.in/gp/bestsellers/books/ref=zg_bs_pg_{i}?pg={i}"
    for i in range(1, 11)
]
```

```python
data = {
    "Rank": [],
    "Title": [],
    "Author": [],
    "Rating": [],
    "Product_URL": []
}
```

```python
for url in urls:
    print("Scraping:", url)

    response = requests.get(url, headers=HEADERS, timeout=15)
    if response.status_code != 200:
        print("Blocked or Failed")
        continue

    soup = BeautifulSoup(response.text, "html.parser")
    items = soup.select("div.zg-grid-general-faceout")

    for item in items:
        # Rank
        rank = item.select_one("span.zg-bdg-text")
        data["Rank"].append(rank.text.strip() if rank else "Not Available")

        # Title
        title = item.select_one("div._cDEzb_p13n-sc-css-line-clamp-1_1Fn1y")
        data["Title"].append(title.text.strip() if title else "Not Available")

        # Author
        author = item.select_one("div.a-row.a-size-small")
        data["Author"].append(author.text.strip() if author else "Not Available")

        # Rating
        rating = item.select_one("span.a-icon-alt")
        data["Rating"].append(rating.text.strip() if rating else "Not Available")

        # Product URL
        link = item.find("a", class_="a-link-normal")
        if link and link.get("href"):
            data["Product_URL"].append("https://www.amazon.in" + link.get("href"))
        else:
            data["Product_URL"].append("Not Available")

    time.sleep(3)

print("✅ Scraping Completed")
```

```python
df = pd.DataFrame(data)
print(df.shape)
df
```
